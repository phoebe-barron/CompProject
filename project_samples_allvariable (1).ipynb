Hypothesis: Higher Chlorophyll A levels will correspond to higher pH levels within a water sample, indicating a possible algal bloom.


```python
#Q1

#imports
import sys
import numpy as np
from scipy import stats
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

file=open(input()) #take input
first= True
LoL=[] #makes new list to append into

#removing header line so it doesn't interfere
for line in file:
    if first==True: #if on the first line; captures headers
        data=line.split(",")
        header=data #saves for later when need to compare diff solutes
        first=False
        
    #once header is removed
    else:
        data=line.split(",") #splits data where there are tabs
        LoL.append(data)

#Q1.a
pH_values=[] #make new lists to append values into
chloroA_values=[]

for row in LoL:
    #grabbing values from proper columns in data and making new lists for them
    pH_values.append(float(row[3]))
    chloroA_values.append(float(row[10]))
                          
#pH stats
meanpH=np.mean(pH_values) #takes the mean of column 4 where pH values are
medpH=np.median(pH_values) #takes median of column 4
stddevpH=np.std(pH_values) #finds the std dev of column 4

#chlorophyll A stats
meanA=np.mean(chloroA_values) #takes the mean of column 4 where pH values are
medA=np.median(chloroA_values) #takes median of column 4
stddevA=np.std(chloroA_values)

#print values
print('The mean pH for all trials is',meanpH)
print('The median pH for all trials is',medpH)
print('The standard deviation of pH for all trials is',stddevpH)

print() #blank line

print('The mean Chlorophyll A levels for all trials is',meanA)
print('The median Chlorophyll A levels for all trials is',medA)
print('The standard deviation of Chlorophyll A levels for all samples is',stddevA)

#Q1.b
#Histogram for pH values
plt.hist(pH_values, bins=5, edgecolor='black')  #5 bins is most useful; tested by changing bin number repeatedly
plt.title('Histogram of pH Values')  #title
plt.xlabel('pH') #labels
plt.ylabel('Frequency')  #labels
plt.show()  #prints plot

#Histogram for Chlorophyll A values
plt.hist(chloroA_values, bins=5, edgecolor='black')  #5 bins is most useful; tested by changing bin number repeatedly
plt.title('Histogram of Chlorophyll A Levels') 
plt.xlabel('Chlorophyll A')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q2

#imports
import sys
import numpy as np
from scipy import stats
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

file=open(input()) #take input
first= True
LoL= [] #makes new list to append into

#removing header line so it doesn't interfere
for line in file:
    if first==True: #if on the first line; captures headers
        data=line.split(",")
        header=data #saves for later when need to compare diff solutes
        first=False
        
    #once header is removed
    else:
        data=line.split(",") #splits data where there are tabs
        LoL.append(data)

pH_values=[] #make new lists to append values into
chloroA_values=[]

for row in LoL:
    #grabbing values from proper columns in data and making new lists for them so stats can be used
    pH_values.append(float(row[3]))
    chloroA_values.append(float(row[10]))
    
#Q2.a

#Pearson correlation
res_p=stats.pearsonr(pH_values, chloroA_values) #takes pearson correlation between pH and chloroA
print(f'Pearson correlation coefficient: {res_p[0]:.3f}') #res.statistic to get coefficient without p-value; 3 decimals

#Spearman correlation
res_s=stats.spearmanr(pH_values, chloroA_values) #takes spearman correlation between pH and chloroA
print(f'Spearman correlation coefficient: {res_s[0]:.3f}') #prints spearman coeff without p value to 3 decimals

#Q2.b

res_lin=stats.linregress(pH_values, chloroA_values) #performs linear regression
print(f"R-squared: {res_lin.rvalue**2:.3f}") #from scipy site: prints R^2 value from linear regression slope to 3 decimal places

#Q2.c
plt.plot(pH_values, chloroA_values, 'o', label='original data')
plt.plot(pH_values, res_lin.intercept + res_lin.slope*np.array(pH_values), 'r', label='fitted line')
#labels
plt.xlabel('pH')  # Label for x-axis
plt.ylabel('Chlorophyll A')  # Label for y-axis
plt.title('pH vs. Chlorophyll A')  #plot title
#showing stuff
plt.legend()
plt.show()
```

```python
#Q3

#imports
import sys
import numpy as np
from scipy import stats
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

file=open(input()) #take input
first= True
LoL= [] #makes new list to append into

#removing header line so it doesn't interfere
for line in file:
    if first==True: #if on the first line; captures headers
        data=line.split(",")
        header=data #saves for later when need to compare diff solutes
        first=False
        
    #once header is removed
    else:
        data=line.split(",") #splits data where there are tabs
        LoL.append(data)

#make new lists to append all variables int for PCA
pH_values = []  #pH
chloroA_values = []  #chlorophyll
DO_values = []  #dissolved oxygen
orp_values = []  #orp
wave630 = []  #individual wavelength readings at different nm
wave647=[]
wave664=[]

for row in LoL:
    #grabbing values from proper columns in data and making new lists for them so stats can be used
    pH_values.append(float(row[3]))
    chloroA_values.append(float(row[10]))
    DO_values.append(float(row[4])) #dissolved oxygen
    orp_values.append(float(row[5])) #orp
    wave630.append(float(row[6]))
    wave647.append(float(row[7]))
    wave664.append(float(row[8]))
    
data = np.array([pH_values, chloroA_values, DO_values, orp_values, wave630, wave647, wave664]).T #prepares data for PCA and flattens wavelength LoL
    
#Q3.a
pca=PCA(n_components=2) #want 2 principle components
#conducting PCA
principle_components=pca.fit_transform(data)
print("Principal components:", principle_components)
print()
print("Explained variance ratio:", pca.explained_variance_ratio_)

#Q3.b
#PC weights
weights=pca.components_
print("Principal component weights:")
print(weights)

#getting data ready for graph
for i in range(weights.shape[0]):
    plt.bar(np.arange(weights.shape[1]) + i * 0.3, weights[i], width=0.3, label=f'PC{i + 1}')

#labels of categories on bar graph (made at 45 angle for readability)
plt.xticks(np.arange(weights.shape[1]), ['pH', 'Chlorophyll A', 'Dissolved Oxygen', 'ORP', 'Wavelength 630', 'Wavelength 647', 'Wavelength 664'],rotation=45, ha='right', fontsize=12)
plt.title('Principal Component Weights')
plt.ylabel('Weight')
plt.legend(title="Principal Components")

# Line for easy visual
plt.axhline(y=0, color='black', linestyle='-')

#color coding components by pH
category = np.array(pH_values)  # Color-code by pH or any other category you want
norm = plt.Normalize(vmin=min(category), vmax=max(category))
cmap = plt.get_cmap('viridis')

#plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(principle_components[:, 0], principle_components[:, 1], c=category, cmap=cmap, norm=norm)

#legend
plt.colorbar(scatter, label='pH')  #colorbar is rainbow

# Add labels and title
plt.xlabel(f'PC1 ({np.round(pca.explained_variance_ratio_[0]*100, decimals=2)}%)')
plt.ylabel(f'PC2 ({np.round(pca.explained_variance_ratio_[1]*100, decimals=2)}%)')
plt.title('Visualization of Data on PC1 and PC2')

plt.show()
```
